### Import libraries

```python
import pandas as pd
import numpy as np
from os import listdir
from os.path import isfile, join
import re
```

### Read the .csv file and display data

```python
Cars= pd.read_csv('/Users/jonathan/Documents/Cars.csv');
Cars.head()
```

### Rename columns

```python
Cars = Cars.rename(columns={'Name': 'Make_and_Model'})
Cars = Cars.rename(columns={'Miles_Away': 'Location'})
Cars = Cars.rename(columns={'Dealership_Name': 'Dealership'})
Cars = Cars.rename(columns={'Phone_Number': 'Phone Number'})
Cars.head()
```

### Split the information in the "Name" (changed to "Make_and_Model") column 

```python
def get_regex_item(name, pattern):
    item = re.search(pattern, name)
    if item is None:
        return None
    else:
        return item.group()
```

```python
Cars['Trim'] = Cars.Make_and_Model.apply(lambda x: x.split()[-1])
Cars['Year'] = Cars.Make_and_Model.apply(lambda x: get_regex_item(x, pattern= r'\d{4}'))
Cars['Condition'] = Cars.Make_and_Model.apply(lambda x: get_regex_item(x, pattern= r'Used'))

Cars.head()
```

### Filter out and rearrange data depending on how important and needed to describe the vehicle

```python
Cars = Cars[['Make_and_Model', 'Year', 'Trim','Condition', 'Mileage', 'Price', 'Dealership']]
Cars.head()
```

### Created a new dataframe 

```python
new_dataframe = pd.DataFrame(
    {
        "Make and Model": ['GMC Yukon', 'GMC Yukon 4WD', 'GMC Yukon 2WD', 'GMC Yukon 2WD', 'GMC Yukon 4X4']
}
)
new_dataframe
```

### Merged the new dataframe and existing dataframe.

```python
Cars2 = pd.concat([Cars, new_dataframe], axis=1, join='inner')
Cars2
```

### Deleted an exisiting column that was no longer needed and rearranged the columns once more

```python
Cars2 = Cars2[['Make and Model', 'Year', 'Trim','Condition', 'Mileage', 'Price', 'Dealership']]
Cars2.head()
```
