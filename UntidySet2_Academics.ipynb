### Split the sex and age of students in two columns

```python
def get_sex_and_age(row):
    sex_and_age = row['sex and age'].split('_')
    sex = sex_and_age[0]
    age = sex_and_age[1]

    sex = 'male' if sex == 'm' else 'female'

    return pd.Series([sex, age])
```
